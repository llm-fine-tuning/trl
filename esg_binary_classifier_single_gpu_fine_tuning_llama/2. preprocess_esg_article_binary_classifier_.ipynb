```python
train_data['label'] = train_data['label'].apply(str)
test_data['label'] = test_data['label'].apply(str)
```

```python
train_data['instruction'] = train_data['content']
train_data['output'] = train_data['text'] + '\n=> ' + train_data['label']
train_data['input'] = ''

test_data['instruction'] = test_data['content']
test_data['output'] = test_data['text'] + '\n=> ' + test_data['label']
test_data['input'] = ''
```

```python
import json

# DataFrame에서 'instruction'과 'output' 열을 딕셔너리 리스트로 변환
train_to_save = train_data[['instruction', 'input', 'output']].to_dict(orient='records')
test_to_save = test_data[['instruction', 'input', 'output']].to_dict(orient='records')

# JSON 파일로 저장
with open('train.json', 'w', encoding='utf-8') as f:
    json.dump(train_to_save, f, ensure_ascii=False, indent=4)

# JSON 파일로 저장
with open('test.json', 'w', encoding='utf-8') as f:
    json.dump(test_to_save, f, ensure_ascii=False, indent=4)
```
