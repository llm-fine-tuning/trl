```python
from tqdm import tqdm
```

```python
response = client.chat.completions.create(
  model="gpt-4o",
  messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": '입력:' + user_prompt + '\n근거:'}
    ]
)
```

```python
print(response.choices[0].message.content)
```

## 반복 호출

```python
result_lst = []

# [:5]를 지우면 전체 호출
for user_prompt in tqdm(df['content'].to_list()[:5]):
  response = client.chat.completions.create(
  model="gpt-4o",
  messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": '입력:' + user_prompt + '\n근거:'}
    ]
)
  result_lst.append(response.choices[0].message.content)
```

```python
result_lst
```

```python
# 주석 처리 제거
df['result'] = result_lst
```

```python
df.to_excel('result.xlsx', index=False)
```
