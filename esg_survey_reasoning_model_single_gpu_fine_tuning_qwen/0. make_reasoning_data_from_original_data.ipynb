```python
from tqdm import tqdm
import pandas as pd
from datasets import load_dataset

# 허깅페이스에서 데이터셋 로드
print("데이터셋 다운로드 중...")
dataset = load_dataset("iamjoon/esg-survey-datasets")

# 데이터셋 정보 확인
print("데이터셋 구조:")
print(dataset)

# train split을 데이터프레임으로 변환 (일반적으로 train split 사용)
if 'train' in dataset:
    df = dataset['train'].to_pandas()
    print(f"\ntrain split 로드 완료: {df.shape}")
else:
    # train split이 없는 경우 첫 번째 split 사용
    split_name = list(dataset.keys())[0]
    df = dataset[split_name].to_pandas()
    print(f"\n{split_name} split 로드 완료: {df.shape}")
```

```python
def change_system_prompt(system_prompt):
  new_systemp_prompt = system_prompt.replace('당신은 주어진 질문에 대해서 검색 결과, 질문, 보기를 바탕으로 답변해야 합니다.\n검색 결과를 바탕으로 답변할 수 없는 경우에는 \'알 수 없음\'이라고 적으세요.', new_added_system_prompt)
  return new_systemp_prompt
```

```python
df['reasoning_system_prompt'] = df['system_prompt'].apply(change_system_prompt)
```

```python
# 변경되었는지 확인
print(df['reasoning_system_prompt'].loc[0][:545])
```

```python
# 5개만 테스트
system_prompts = df['reasoning_system_prompt'].to_list()
user_prompts = df['user_prompt'].to_list()

result_lst = []
for system_prompt, user_prompt in tqdm(zip(system_prompts[:5], user_prompts[:5]), total=len(df[:5])):
    response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
      {"role": "system", "content": system_prompt},
      {"role": "user", "content": user_prompt}
    ],
    temperature=0
    )
    result_lst.append(response.choices[0].message.content)
```
