```python
result_lst = []

for user_prompt in tqdm(df['content'].to_list()[3000:]):
  response = client.chat.completions.create(
  model="gpt-4o",
  messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": '입력:' + user_prompt + '\n근거:'}
    ]
)
  result_lst.append(response.choices[0].message.content)
```

```python
# 주석 처리 제거
df.loc[3000:, 'result'] = result_lst
```

```python
# result 필드를 "=>" 앞뒤로 나누어 'text'와 'label' 필드로 구분하여 저장
df[['score_text', 'score']] = df['result'].str.split('=>', expand=True)

# label 필드에 포함된 공백을 제거하고 'true' 또는 'false'로 변환
df['score'] = df['score'].str.strip().str.lower()
```

```python
df.to_excel('result.xlsx', index=False)
```
